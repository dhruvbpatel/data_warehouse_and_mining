```python
def checkRange(mean,std,col): ## function to check if data in each column is noraml by checking its range in mean+std & mean-std
    maxRange = mean+std
    minRange = mean-std
    coldata  = data.loc[:,col]     ## each column data 
    
    f_count=0
    s_count=0
    t_count=0
    
    for i in range(len(data)):
        if(coldata[i] <= maxRange and coldata[i] >=minRange):
            f_count+=1
        if(coldata[i]<=maxRange+std and coldata[i] >=minRange-std):
            s_count+=1
        if(coldata[i]<=maxRange+2*std and coldata[i] >=minRange-2*std):
            t_count+=1
        
    f_per = (f_count*100)/len(data)
    s_per = (s_count*100)/len(data)
    t_per= (t_count*100)/len(data)
        
    print(f_per," ",s_per," ",t_per)
```

```python
# passing mean, std, and column name  this function will give percentage dist for all data

for i in range(len(data.columns)):
    mn = data.mean()[i]
    std = data.std()[i]
    checkRange(mn,std,data.columns[i])

```

### From this calculation we can observe that each variables(columns) are normally distributed and thus we can say the data given is as a whole , normally distributed
