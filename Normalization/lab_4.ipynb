```python
## Min-Max Normalization

def minMaxScaler(data,col):
    
    minData = float(data.min()[col])
    maxData = float(data.max()[col])
    
    for i in range(len(data.iloc[:,col])):
        curr_val = float(data.iloc[:,col][i])
        
        normalize = float(curr_val-minData)/(maxData-minData)
        
        data.iloc[:,col][i]=normalize   
        
    return data
    
    
```

```python
for i in range(len(df.columns)):
    df.iloc[:,i] = minMaxScaler(df.astype(float),i)

```

```python
def zscore_scaler(col):
    
    mean=float(col.mean())
    std = float(col.std())
    col = col.astype(float)
    for i in range(len(col)):
        col[i]=(col[i]-mean)/std

    return col

        
    
    
    
```

```python
for i in df.columns:
    df[i]=zscore_scaler(df[i])

```

```python
def decimanl_norm(col):
    col = col.astype(float)
    max_num = float(col.max())
    powr= len(str(max_num))
    
    for i in range(len(col)):
        col[i]=col[i]/pow(10,powr)

    return col

    
    
```

```python
for i in df.columns:
    df[i]=decimanl_norm(df[i])
        
```
